```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(my_data_pandas['Height'], my_data_pandas['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Comparing Height and Weight')
```

```python
import seaborn as sns
```
